```python
from datasets import load_dataset
from collections import defaultdict
from PIL import Image
import os
```

```python
dataset = load_dataset(
    "detection-datasets/coco",
    split="train",
    streaming=True
)
```

```python
TARGET_CLASSES = [
    "person",
    "car", "truck", "bus", "motorcycle", "bicycle", "airplane",
    "traffic light", "stop sign", "bench",
    "dog", "cat", "horse", "bird", "cow", "elephant",
    "bottle", "cup", "bowl", "pizza", "cake",
    "chair", "couch", "bed", "potted plant"
]
```

```python
class_count = defaultdict(int)
MAX_PER_CLASS = 100
filtered_samples = []

for sample in dataset:
    labels = sample["objects"]["label"]   # 👈 LIST of labels

    for label in labels:
        if label in TARGET_CLASSES and class_count[label] < MAX_PER_CLASS:
            filtered_samples.append(sample)
            class_count[label] += 1
            break

    if all(class_count[c] >= MAX_PER_CLASS for c in TARGET_CLASSES):
        break

```

```python
os.makedirs("data/raw/images", exist_ok=True)
```

```python
for i, sample in enumerate(filtered_samples):
    img = sample["image"]
    img.save(f"data/raw/images/img_{i}.jpg")
```
